```python
import nltk
```

```python
nltk.download()
```

```python
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
```

```python
def remove_punc(read):
  for ele in read:
    if ele in punc:
        read = read.replace(ele, " ")
  return read
```

```python
tokens = {}

for folder in folders:
    folder_path = os.path.join(directory_path, folder)
    all_files = [file for file in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, file))]

    # For each random number, create a file name and process the file if it exists
    for i in random_numbers:
        file_name = folder + '_' + str(i) + '.txt'
        file_path = os.path.join(folder_path, file_name)
        if os.path.exists(file_path):
            with open(file_path, 'r') as file:
                data = file.read()
                data = remove_punc(data)
                token = nltk.word_tokenize(data)
                token = [word.lower() for word in token if not word.isdigit()]
                unique_tokens = list(set(token))
                tokens[file_name] = unique_tokens
```

```python
tokens
```

```python
inverted_index = {}

for doc_name,doc_tokens in tokens.items():
  for token in doc_tokens:
    if token not in inverted_index:
      inverted_index[token] = {doc_name}
    else:
      inverted_index[token].add(doc_name)
```

```python
for token in inverted_index:
    inverted_index[token] = sorted(inverted_index[token], key=lambda x: int(x.split('_')[1].split('.')[0]))
```

```python
def extract_doc_number(doc_name):
    return int(doc_name.split('_')[1].split('.')[0])
```

```python
def and_postings(posting1, posting2):
    i = 0
    j = 0
    result = []
    while i < len(posting1) and j < len(posting2):
        doc1 = extract_doc_number(posting1[i])
        doc2 = extract_doc_number(posting2[j])
        if doc1 == doc2:
            result.append(posting1[i])
            i += 1
            j += 1
        elif doc1 < doc2:
            i += 1
        else:
            j += 1
    return result
```

```python
random_tokens = random.sample(all_tokens, 2)

query1 = list(inverted_index.get(random_tokens[0]))
query2 = list(inverted_index.get(random_tokens[1]))
res = and_postings(query1, query2)
res
```
