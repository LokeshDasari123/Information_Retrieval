```python
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
```

```python
def remove_punc(read):
  for ele in read:
    if ele in punc:
        read = read.replace(ele, " ")
  return read
```

```python
tokens = {}
for i in random_numbers:
    file_name = 'business_' + str(i) + '.txt'
    file_path = '/content/drive/MyDrive/CSE 419/Lab Assign-2/Documents/business/' + file_name
    with open(file_path, 'r') as file:
        data = file.read()
        data = remove_punc(data)
        token = nltk.word_tokenize(data)
        token = [word for word in token if not word.isdigit()]
        unique_tokens = list(set(token))
        tokens[file_name] = unique_tokens
```

```python
inverted_index = {}

for doc_name,doc_tokens in tokens.items():
  for token in doc_tokens:
    if token not in inverted_index:
      inverted_index[token] = {doc_name}
    else:
      inverted_index[token].add(doc_name)
```

```python
inverted_index
```

```python
for token in inverted_index:
    inverted_index[token] = sorted(inverted_index[token], key=lambda x: int(x.split('_')[1].split('.')[0]))
```

```python
for term, doc_list in inverted_index.items():
    print(f"{term}: {doc_list}")
```
