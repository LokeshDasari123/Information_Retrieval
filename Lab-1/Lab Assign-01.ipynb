```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
tokens1_without_stopwords = [word for word in tokens_1 if word.lower() not in stop_words]
token_count_1 = len(tokens1_without_stopwords)
print(token_count_1)
```
