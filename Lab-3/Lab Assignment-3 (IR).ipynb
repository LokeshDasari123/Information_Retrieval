```python
selected_docs_path = []
selected_filenames = []

for folder in folders:
  folder_path = os.path.join(dataset_path,folder)
  docs = os.listdir(folder_path)
  selected_docs = [os.path.join(folder_path, doc) for doc in random.sample(docs, 10)]
  selected_docs_path.extend(selected_docs)
  selected_filenames.extend([os.path.basename(doc) for doc in selected_docs])
```

```python
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
```

```python
def remove_punc(read):
    return read.translate(str.maketrans('', '', punc))
```

```python
from collections import defaultdict
```

```python
inverted_index = defaultdict(lambda: defaultdict(list))
```

```python
for doc_id, (doc_path, doc_name) in enumerate(zip(selected_docs_path, selected_filenames)):
    if os.path.exists(doc_path):
        with open(doc_path, 'r') as file:
            data = file.read()
            data = remove_punc(data)
            tokens = nltk.word_tokenize(data)
            tokens = [word.lower() for word in tokens]

            # Build the inverted index
            for pos, token in enumerate(tokens):
                inverted_index[token][doc_id].append(pos)
```

```python
def execute_phrasal_query(phrase, inverted_index):
    tokens = phrase.lower().split()  # Tokenize the phrase
    if not tokens:
        return []

    # Find document IDs where the phrase may be present
    doc_ids = set(inverted_index[tokens[0]].keys())

    # For each subsequent token, refine the set of document IDs
    for token in tokens[1:]:
        if token not in inverted_index:
            return []  # If token not in index, phrase cannot be found
        doc_ids &= set(inverted_index[token].keys())

    result_docs = []

    for doc_id in doc_ids:
        positions = [inverted_index[token][doc_id] for token in tokens]
        # Check if all tokens are present consecutively in the document
        if check_consecutive_positions(positions):
            result_docs.append(doc_id)

    return result_docs
```

```python
def check_consecutive_positions(positions):
    # Flatten the list of positions and sort
    flat_positions = sorted([pos for sublist in positions for pos in sublist])

    # Check for consecutive positions
    return all(flat_positions[i] + 1 == flat_positions[i + 1] for i in range(len(flat_positions) - 1))
```

```python
phrase = "morocco mauritius rwanda senegal"
result = execute_phrasal_query(phrase, inverted_index)
print("Documents containing the phrase:", result)
```
